# Etude théorique de la formation d'un nuage
La formation d'un nuage résulte de la condensation de l'eau présente dans l'air à une certaine altitude.
D'un point de vue thermodynamique, la condition un nuage se forme lorsque la pression réelle de l'eau $P_e$ devient supérieur à la pression saturante $P_{sat}$ de l'eau dans l'air.  
  
Pour pouvoir entamer cette étude il faut faire un certain nombre d'hypothèse et d'approximation.
- L'air est un milieu très mauvais conducteur de chaleur, si bien que nous considérerons que toutes les transformations y ayant lieu sont adiabatique $\delta Q = 0$
- L'air est considéré comme un gaz parfait.
- L'air est approximé a un fluide parfait, viscosité nulle $\mu = 0$
- L'écoullement de l'air est irrotationnel $\overrightarrow{\nabla}\land\overrightarrow{v}=\overrightarrow{0}$
- L'écoullement est supposé $\frac{\partial\overrightarrow{v}}{\partial t}=0$
- L'air est quasiment incompressible $\frac{\partial\rho}{\partial t}=0$
  
Commençons par décrire les mouvements de l'air à proximité de la montagne. Pour cela, on énonce l'équation de Navier-Stokes :
$$\rho\left(\frac{\partial\overrightarrow{v}}{\partial t}+\left(\overrightarrow{v}\cdot\overrightarrow{\nabla}\right)\overrightarrow{v}\right)=-\overrightarrow{\nabla}P+\mu\overrightarrow{\Delta}\overrightarrow{v}+\overrightarrow{f_{ext}}$$
Par les hypothèses considérées on obtient :
$$\rho\frac{\partial\overrightarrow{v}}{\partial t}+\rho\left(\overrightarrow{v}\cdot\overrightarrow{\nabla}\right)\overrightarrow{v}=-\overrightarrow{\nabla}P+\rho\overrightarrow{g}$$
On peut réécrire l'accélération convective, en utilisant la forme dite de "Lamb" :
$$\left(\overrightarrow{v}\cdot\overrightarrow{\nabla}\right)\overrightarrow{v}=\frac{1}{2}\overrightarrow{\nabla}\|\overrightarrow{v}\|^2-\overrightarrow{v}\land\left(\overrightarrow{\nabla}\land\overrightarrow{v}\right)$$
Or nous avons supposés que l'écoulement est irrotationnel, donc :
$$\left(\overrightarrow{v}\cdot\overrightarrow{\nabla}\right)\overrightarrow{v}=\frac{1}{2}\overrightarrow{\nabla}\|\overrightarrow{v}\|^2$$
On a donc :
$$\rho\frac{\partial\overrightarrow{v}}{\partial t}+\frac{\rho}{2}\overrightarrow{\nabla}\|\overrightarrow{v}\|^2=-\overrightarrow{\nabla}P+\rho\overrightarrow{g}$$
De plus, l'écoulement est stationnaire, on obtient :
$$\frac{1}{2}\overrightarrow{\nabla}\|\overrightarrow{v}\|^2=-\frac{1}{\rho}\overrightarrow{\nabla}P+\overrightarrow{g}$$
Or le poids est une force conservative, on peut donc l'écrire sous la forme d'un gradient de potentiel. Sachant qu'avec l'équation de Navier-Stockes nous utililons des forces par unité de volume, nous avons :
$$\overrightarrow{P}=-\overrightarrow{\nabla}gz$$
Donc :
$$\frac{1}{2}\overrightarrow{\nabla}v^2+\frac{1}{\rho}\overrightarrow{\nabla}P+\overrightarrow{\nabla}gz=0$$
$$\Leftrightarrow\overrightarrow{\nabla}\left(\frac{v^2}{2}+\frac{P}{\rho}+gz\right)=0$$
$$\Leftrightarrow\frac{v^2}{2}+\frac{P}{\rho}+gz=B$$
Où $B$ est la constante de Bernouilli.  
  
Pour utiliser cette equation il nous en faut une seconde qui décris l'évolution de la pression. On utilise la thermodynamique pour cela.
Pour décrire les transformation thermodynamique de l'air, on choisi une fonction d'état. Dans notre cas, l'enthalpie $H$ est bien adaptée.
$$H=U+PV$$
$$dH=dU+PdV+VdP$$
$$dH=\delta Q - PdV+PdV+VdP$$
$$dH =  VdP $$
$$dH = \frac{m}{\rho} dP$$
Sachant cela, on écrit l'enthalpie massique :
$$dh = \frac{1}{\rho} dP$$
L'enthalpie nous donne une valeur pour la pression : $h=\frac{P}{\rho}$.
Dans ce cas la constante de Bernouilli devient :
$$B = \frac{v^2}{2}+h+gz$$
  
Nous souhaitons déterminer une expression de l'enthalpie de l'air. On reprend son expression intégrale :
$$H=U+PV$$
Or ici l'air est considéré comme un gaz parfait, donc l'energie interne de ce dernier ne dépend que de la température. On donne :
$$U=\frac{5}{2}nRT$$
L'enthalpie vaut donc :
$$H=U+PV$$
$$\Leftrightarrow H=\frac{5}{2}nRT+PV$$
Or $PV=nRT$, donc :
$$\Leftrightarrow H=\frac{5}{2}nRT+nRT=\frac{7}{2}nRT$$
L'enthalpie massique de l'air vaut donc :
$$h=\frac{H}{m_{air}}=\frac{7nRT}{2m_{air}}=\frac{7}{2}\frac{RT}{M_{air}}$$
On se sert de cette relation pour déterminer la constante de Bernouilli et également pour trouver la tempérture $T$ en un point connaissant son enthalpie $h$.
Loin de la montagne (à l'infini), la constante de Bernouilli $B$ vaut :
$$B=\frac{v_{in}^2}{2}+\frac{7}{2}\frac{RT_0}{M_{air}}$$
On prendra $T_0 = 283$ K et $v_{in}=20$ m/s, nous obtenons : $B=284509,2353$.  
  
Toutes ces équations nous permettent de déterminer la température d'un point connaissant la vitesse de l'écoulement :

$$T(h)=\frac{2}{7}\frac{hM_{air}}{R}$$
$$h(v)=B-\left(\frac{v^2}{2}+gz\right)$$

La pression saturante $P_sat$ est une grandeur entièrement conditionnée par la température. Elle est donnée par : 
$$P_{sat}(T)=\left(\frac{T-273}{40}+1\right)^2\times 10^3$$
Quant à la pression partielle de l'eau dans le milieu, elle entièrement régie par la pression au point considéré.
$$P_e(P)=\frac{rP}{\frac{M_{eau}}{M_{air}}+r}$$
L'air étant considéré comme un gaz parfait et mauvais conducteur de chaleur, on peut utiliser la Loi de Laplace pour déterminer la pression.
$$P_0T_0^{-\frac{\gamma}{\gamma-1}}=PT^{-\frac{\gamma}{\gamma-1}}$$
On prend pour exposant : $\frac{\gamma}{\gamma-1}=\frac{7}{2}$.On a donc :
$$P(T)=P_0T_0^{-\frac{7}{2}}\times T^{-\frac{7}{2}}$$
  
  
Nous avons toutes les équation nécéssaires à la thermodynamique du nuage. Mais toutes ces équation dépendent d'un paramètre sur lequel nous n'avons encore aucune information :lLa vitesse $v$.
  
On détermine cette dernière à partir de l'équation de continuité :
$$\frac{\partial\rho}{\partial t}+\overrightarrow{\nabla}\cdot\rho\overrightarrow{v}=0$$
$$\Leftrightarrow \rho\overrightarrow{\nabla}\cdot\overrightarrow{v}=0$$
On suppose que l'écoullement à lieu dans un plan vertical $(x,z)$. On a donc :
$$\overrightarrow{\nabla}\cdot\overrightarrow{v}=\frac{\partial v_x}{\partial x}+\frac{\partial v_z}{\partial z}=0$$
$$\Leftrightarrow \frac{\partial v_x}{\partial x}=-\frac{\partial v_z}{\partial z}$$

Il existe donc une fonction appelée fonction de courant $V(x,z)$ telle que :
$$v_x=\frac{\partial V}{\partial z}$$
$$v_z=\frac{\partial v_z}{\partial x}$$

De plus l'écoullement étant irrotationnel, on a :
$$\overrightarrow{\nabla}\land\overrightarrow{v}=\frac{\partial v_z}{\partial x}-\frac{\partial v_x}{\partial z}=0$$
$$\Leftrightarrow \frac{\partial^2 V}{\partial x^2}+\frac{\partial^2 V}{\partial z^2} = 0$$
$$\Leftrightarrow \Delta V = 0$$

La vitesse $v$ est donc donnée par les dérivées de la fonction de courant $V$.  
  
Nous avons donc désormais toutes les expressions et calculs nécéssaires pour la simulation de la création d'un nuage aux abords d'une montagne.

## En résumé
- La fonction de courant $V$ nous permet de connaître la vitesse  
- La constante de Bernouilli $B$ nous permet d'évaluer l'enthalpie $h$ en tout points à partir des vitesses et de la hauteur.
- On se sert de la deuxième formule de l'enthalpie $h$ pour déterminer la température $T$ en tout points.
- On peut déterminer la pression réelle $P$ en servant de la Loi de Laplace.
- On évalue les pressions saturante $P_{sat}$ et partielle de l'eau $P_e$

# Programme :
Dans un premier temps, nous importons les modules nécessaires à l'utilisation du programme

```python
## Import des modules nécessaires ##

import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

On initialise les données et constantes utiles à la simulation

```python
## Donnees du problème ##

M_eau = 0.01801528 #kg/mol
M_air = 0.028965 #kg/mol
v_in = 72 #km/h
T_0 = 283 #K
P_0 = 10**5
R = 8.314 #KJ/mol
g = 9.81
B = v_in**2/2+7/2*(R*T_0)/M_air
```

```python
def courant_init(v_max, dim, exp = False):
    mat_courant = np.zeros((dim, dim))
    v_j = v_max / dim
    for j in range(dim):
        mat_courant[dim - 1 - j,0] = v_j
        mat_courant[dim - 1 - j,dim - 1] = v_j
        v_j += v_max / dim
    for i in range(dim):
        mat_courant[0,i] = v_max
    return mat_courant

## Méthodes mathématiques & numériques ##

def laplacien_relax(matrice, montagne):
    dim = matrice.shape[0]
    x = sp.symbols('x')
    for b in range(25):
        for i in range(1, dim - 1):
            a=dim-1-i
            for j in range(1, dim - 1):
                    if a<=montagne.subs(x, j):
                        if i<= dim//2:
                            matrice[a,j] = 0
                    else:
                        matrice[i,j] = 1/4 * (matrice[i - 1,j] + matrice[i + 1,j] + matrice[i,j - 1] + matrice[i,j + 1]-4*matrice[i,j])
    matrice[j+1]=0
    return matrice

def v_x(mat) :
    a = mat.shape[0]
    mat_v_x = np.zeros((a-2,a-2))
    for i in range(0,mat_v_x.shape[0]) :
        for j in range(0,mat_v_x.shape[0]) :
            mat_v_x[i,j] = (mat[i,j]-mat[i,j+2])/2
    return mat_v_x

def v_z(mat) :
    a = mat.shape[0]
    mat_v_x = np.zeros((a - 2,a - 2))
    for i in range(0, mat_v_x.shape[0]):
        for j in range(0, mat_v_x.shape[0]):
            mat_v_x[j,i] = -(mat[j,i] - mat[j,i + 2]) / 2
    return mat_v_x
```

```python
## Condition thermodynamique de la formation du nuage ##

def enthalpie(v_1,v_2,z) :
    return B-((v_1**2+v_2**2)/2+g*z)

def Temp(h) :
    return (h*M_air)/R*2/7

def Pr(T) :
    return T**(7/2)*P_0*T_0**(-7/2)

def P_sat(T) :
    return ((T-273)/40+1)*10**3

def P_e(r,T,P) :
    return (r*P)/(M_eau/M_air+r)

def mat_P_sat(v_x,v_z) :
    mat = np.zeros((v_x.shape[0],v_x.shape[0]))
    for i in range(v_x.shape[0]) :
        for j in range(v_z.shape[0]):
            mat[i,j] = P_sat(Temp(enthalpie(v_x[i,j],v_z[i,j],v_x.shape[0]-j-1)))
    return mat
    
def mat_P_e(r,v_x,v_z) :
    mat = np.zeros((v_x.shape[0],v_x.shape[0]))
    for i in range(v_x.shape[0]) :
        for j in range(v_x.shape[0]) :
            mat[i,j] = P_e(r,Temp(enthalpie(v_x[i,j],v_z[i,j],v_x.shape[0]-j-1)),Pr(Temp(enthalpie(v_x[i,j],v_z[i,j],v_x.shape[0]-j-1))))
    return mat

def pititnuage(r,v_x,v_z) :
    mat_partielle = mat_P_e(r,v_x,v_z)
    mat_saturante = mat_P_sat(v_x,v_z)
    return mat_partielle > mat_saturante
```

```python
## Fonctions d'affichage ##

def ec_mont(ecran, montagne):
    for j in range(ecran.shape[1]):
        for i in range(ecran.shape[0]):
            if i < montagne.subs('x', j):
                ecran[ecran.shape[0] - 1 - i,j] = 1
    return ecran

#def masque(mat_cd,montagne) :
#    # masque[mat_cd] = 2
#    return masque

def affiche_matrice(ecran):
    lignes, colonnes = ecran.shape
    cases_nulles=ecran==0
    couleurs=np.zeros((lignes, colonnes, 3))
    couleurs[cases_nulles]=[0,0,0]
    couleurs[~cases_nulles]=[0.5,0.7,1.0]        
    ecran=plt.imshow(couleurs)
    ecran=plt.axis('off')
    return ecran

def plot_binary_matrix(boolean_matrix):
    # Convertir la matrice de booléens en une matrice d'entiers (0 pour False, 1 pour True)
    int_matrix = np.array(boolean_matrix, dtype=int)

    # Créer un graphique en utilisant imshow pour afficher la matrice en noir et blanc
    plt.imshow(int_matrix, cmap='binary', interpolation='nearest')

    # Masquer les graduations sur les axes
    plt.xticks([])
    plt.yticks([])

    # Afficher le graphique
    plt.show()
```

```python
## Paramètres modifiables par l'utilisateur ##

dim = 100 #Dimension de la matrice (précision de la discrétisation)
v = 10 #Répartition des vitesses selon l'altitude (Fonction Sympy en z)
montagne = sp.sympify(f"-(x/4-{dim}/8)**2+2*{dim}/3") #Forme de la montagne selon la position horizontale (Fonction Sympy en x)
r = 1/100 #Taux d'humidité de l'air en fonction de l'altitude et du temps (Fonction Sympy en z et t)
```

```python
## Programme principal ##

mat_courant = courant_init(v,dim)
mat_courant = laplacien_relax(mat_courant, montagne)
v_1 = v_x(mat_courant)
v_2 = v_z(mat_courant)
cd = pititnuage(r,v_1,v_2)
a = plot_binary_matrix(cd)
```
